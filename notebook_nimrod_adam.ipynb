# Exercice 1 : MAC ins√©cure

```python
import random

def generate_k_bit_key(k):
    key = random.getrandbits(k)
    key_str = format(key, '0' + str(k) + 'b')
    return key_str

def xor_k_bit_key_and_message(key, message):
    if len(key) != len(message):
        raise ValueError("k and m must be same length")
    result = ''.join('1' if key[i] != message[i] else '0' for i in range(len(key)))
    return result

def verify(tag, key, message):
    return tag == xor_k_bit_key_and_message(key, message)

# Messages
m1 = '0' * 32
m2 = '1' * 32
m3 = '0' * 16 + '1' * 16

messages = [m1, m2, m3]

# Repeat the scenario three times
for i in range(3):
    print(f"Scenario {i+1}:")
    
    # Generate a 32-bit key
    k = generate_k_bit_key(32)
    print(f"Alice and Bob's key (k): {k}")
    
    for m in messages:
        # Alice generates the tag
        t = xor_k_bit_key_and_message(k, m)
        print(f"Alice's message (m): {m}")
        print(f"Alice's tag (t): {t}")
        
        # Eve passes the message and tag to Bob (no modification in this scenario)
        received_m = m
        received_t = t
        print(f"Eve's passes message (m): {received_m}")
        print(f"Eve's passes tag (t): {received_t}") 
        # Bob verifies the message
        v = verify(received_t, k, received_m)
        print(f"Bob's received message (m): {received_m}")
        print(f"Bob's received tag (t): {received_t}")
        print(f"Bob's verification result (v): {v}")
        print()
```
